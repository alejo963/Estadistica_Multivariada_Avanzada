# Solución por descenso por el gradiente de la regresión logística

En este cuaderno se explora la solución por gradiente descendiente de la regresión logística.

**Predicciones:**
$$
\hat{y} =
  \begin{cases}
    1       & \quad \text{si } \sigma(\mathbf{x}^T\mathbf{w}) \ge 0.5\\
    0  & \quad \text{si } \sigma(\mathbf{x}^T\mathbf{w}) < 0.5
  \end{cases}
$$

También podemos usar simplemente $\sigma(\mathbf{x}^{T}\mathbf{w})$ si queremos conservar la probabilidad de que una observación $\mathbf{x}$ pertenezca a una clase en particular.

**Función de costo:**
$$
-\frac{1}{n}\sum_{i=1}^{n}y_i\ln(\hat{y}_i) + (1-y_i)\ln(1-\hat{y}_i)
$$

**Gradiente de la función de costo:**
$$
\frac{1}{m}X^{T}(\hat{y}-y)
$$
**Solución por gradiente descendiente:**
```
repetir hasta criterio de parada {
    ...
    dw = ...
    w := w - alpha*dw
}
```

````python
def plot_binary_classification(X, y, w=None):
    """ Make a plot of X, y data.
  
    Args:
        X (numpy.array): input variables
        y (numpy.array): classes
        w (numpy.array): weights
    """
    cmap = ListedColormap(['#FF0000', '#0000FF'])
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 0.2, X[:, 0].max() + 0.2
    x2_min, x2_max = X[:, 1].min() - 0.2, X[:, 1].max() + 0.2
    
    xx1, xx2 = np.meshgrid(
        np.arange(x1_min, x1_max, 0.01), np.arange(x2_min, x2_max, 0.01)
    )
    
    if w is not None:
        X_ = np.array([xx1.ravel(), xx2.ravel()]).T
        X_ = np.concatenate((np.ones((X_.shape[0], 1)), X_), axis=1)
    
        z = compute_predictions(X_, w)
        z = np.reshape(z, xx1.shape)
        plt.contourf(xx1, xx2, z, alpha=0.3, cmap=cmap)
    
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    plt.scatter(X[:,0], X[:,1], c=y, cmap=cmap)
````

## Programación de gradiente descendiente

En este ejercicio vamos a implementar gradiente descendiente como solución al modelo conocido como **regresión logística**. Para esto es importante implementar varias funciones que nos permitan calcular el gradiente, la función de costo, etc. 

### 1. Sigmoide
Implementa la función `sigmoid(z)`. Esta función debe funcionar para cuando `z` es un arreglo y no simplemente un escalar.

````python
def sigmoid(z):
    # --- YOUR CODE STARTS HERE
    activation = 1 / (1 + np.exp(-z))
    # --- YOUR CODE ENDS HERE
    return activation
````

````python
fig = plt.figure(figsize=[7, 5])
ax = fig.add_subplot(111)
ax.set_axisbelow(True)
ax.grid(linestyle='--',color='0.7')
ax.set_xlim(-10, 10)
ax.plot(np.linspace(-10, 10, num=10000), sigmoid(np.linspace(-10, 10, num=10000)), color='g', alpha=0.65)
````

### 2. Predicciones
Implementa la función que permite hacer predicciones usando la regresión logística. Incluye un argumento adicional que permita retornar las probabilidades de que los inputs pertenezcan a la clase 1

````python
def compute_predictions(X, w, return_probs=False):
    # --- YOUR CODE STARTS HERE
    probs = sigmoid(X @ w)
    # --- YOUR CODE ENDS HERE
    if return_probs:
        return probs
    else:
        # --- YOUR CODE STARTS HERE
        preds = (probs >= 0.5) * 1
        # --- YOUR CODE ENDS HERE
        return preds
````

### 3. Costo
Implementa la función `compute_cost(y, y_pred)` , la cual calcula el costo de las predicciones actuales del modelo de regresión logística.

````python
1- np.array([1,2,3])
````

````python
def compute_cost(y, y_pred):
    # --- YOUR CODE STARTS HERE
    cost = -np.mean(y * np.log(y_pred)+ (1 - y)*np.log(1-y_pred))
    # --- YOUR CODE ENDS HERE
    return cost
````

### 4. Gradientes
Implementa la función `compute_gradient(X, y, y_pred)`, la cual es necesaria para la implementación de gradiente descendiente.



````python
def compute_gradient(X, y, y_pred):
    # --- YOUR CODE STARTS HERE
    grad = 1 / len(y) * X.T @ (y_pred - y)
    # --- YOUR CODE ENDS HERE 
    return grad
````

### 5. Optimización
Implementa la función `optimize_logistic_regression(X, y, learning_rate, num_iterations)`, la cual calcula los parámetros de la regresión logística. En esta ocación, el criterio de parada será simplemente un número de iteraciones. Para propositos ilustrativos, guarda en una lista el costo y los pesos en cada iteración

````python
def optimize_logistic_regression(X, y, learning_rate=0.1, num_iterations=1000):
    
    y = np.reshape(y, newshape=(-1, 1))
    X = np.concatenate((np.ones((y.shape[0], 1)), X), axis=1)
    
    num_points, num_features = X.shape
    
    rgen = np.random.RandomState(42)
    w = rgen.normal(loc=0.0, scale=0.01, size=(num_features, 1))
    
    costs = []
    weights = []
  
    for it in range(1, num_iterations+1):
        # --- YOUR CODE STARTS HERE
        y_pred = compute_predictions(X, w, True)
                
        cost = compute_cost(y, y_pred)
        
        dw = compute_gradient(X, y, y_pred)
        
        w = w - learning_rate * dw
        # --- YOUR CODE ENDS HERE
        costs.append(cost)
        weights.append(w)
    return weights, costs
````

````python
weights, costs = optimize_logistic_regression(X, y)
````
