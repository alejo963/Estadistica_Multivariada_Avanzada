Hay que hacer preparación y limpieza de datos malformados. Primero, algunas variables numéricas terminan con guión bajo: Eliminemoslo

```python
df.Age = df.Age.str.replace("_","")
```

```python
df.Age = df.Age.astype('int')
```

```python
df.Annual_Income = df.Annual_Income.str.replace("_","").astype('float')
```

```python
df.info()
```

```python
df.Num_of_Loan = df.Num_of_Loan.str.replace("_","").astype('int')
```

```python
df.Num_of_Delayed_Payment
```

```python
df.Num_of_Delayed_Payment = df.Num_of_Delayed_Payment.str.replace("_", "").astype('float')
```

```python
df.loc[df.Changed_Credit_Limit == "_", 'Changed_Credit_Limit'] = np.NaN
```

```python
df.Changed_Credit_Limit = df.Changed_Credit_Limit.str.replace("_", "").astype('float')
```

```python
df.Outstanding_Debt = df.Outstanding_Debt.str.replace("_", "").astype('float')
```

```python
df.Amount_invested_monthly = df.Amount_invested_monthly.str.replace("_", "").astype('float')
```

```python
df.Monthly_Balance = df.Monthly_Balance.str.replace("_", "").astype('float')
```
